3) В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. 
Какова вероятность того, что все извлеченные детали окрашены?

```python
# Число вариантов извлечения 3х из 15и деталей (с учетом порядка) - размещений:          15!/(15-3)! = 15*14*13
# Число вариантов извлечь 3 детали, оказавшиеся покрашенными из 15и (с учетом порядка):  9*8*7
# Искомая величина вероятности =
print((9*8*7)/(15*14*13))
```
